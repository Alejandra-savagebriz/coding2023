# Pandas

# Filter
**Exercise 1: Filter Late Flights from JFK**
Filter the flights that departed from JFK Airport ('JFK') and had a departure delay greater than 30 minutes. (15241 rows)




```python
# @title
df1 = flights[(flights['origin'] == 'JFK') & (flights['dep_delay'] > 30)]
print(df1.shape)
```

**Exercise 2: Filter Long Flights with Arrival Delays**
Filter flights with a distance greater than 1500 miles and an arrival delay greater than 15 minutes.(15553 rows)



```python
# @title
df2 = flights[(flights['distance'] > 1500) & (flights['arr_delay'] > 15)]
print(df2.shape)
```




**Exercise 3: Filter American Airlines (AA) Flights with Short Airtime**
Filter American Airlines (carrier code 'AA') flights with an airtime less than 2 hours.(5357 rows)





```python
# @title
df3 = flights[(flights['carrier'] == 'AA') & (flights['air_time'] < 120)]
print(df3.shape)
```

**Exercise 4: Filter Early Morning Flights with Long Delays**
Filter flights that departed between 4 AM and 6 AM (hour >= 4 and hour <= 6) and had a departure delay greater than 60 minutes.(602 rows)

```python
# @title
df4 = flights[(flights['hour'] >= 4) & (flights['hour'] <= 6) & (flights['dep_delay'] > 60)]
print(df4.shape)
```

# Groupby aggragate
Exercise 1: Average Departure Delay by Carrier
Calculate the average departure delay and the total number of flights for each carrier.



```python
flights.head()
```

```python
# @title
carrier_stats = flights.groupby('carrier').agg({'dep_delay': 'mean', 'flight': 'count'}).reset_index()
carrier_stats.rename(columns={'dep_delay': 'avg_dep_delay', 'flight': 'total_flights'}, inplace=True)
print(carrier_stats.shape)
carrier_stats.head()
```

Exercise 2: Maximum Arrival Delay by Month
Find the maximum arrival delay for each month in the dataset.

```python
# @title
monthly_max_arr_delay = flights.groupby('month')['arr_delay'].max().reset_index()
print(monthly_max_arr_delay.shape)
monthly_max_arr_delay.head()
```

Exercise 3: Calculate the average flight distance for each combination of origin airport ('origin') and carrier ('carrier').

```python
# @title
avg_distance_by_origin_carrier = flights.groupby(['origin', 'carrier'])['distance'].mean().reset_index()
print(avg_distance_by_origin_carrier.shape)
avg_distance_by_origin_carrier.head()
```

Exercise 4: Total Departure Delays by Hour and Origin
Find the total departure delays for each hour of the day and each origin airport.



```python
# @title
total_dep_delays_by_hour_origin = flights.groupby(['hour', 'origin'])['dep_delay'].sum().reset_index()
print(total_dep_delays_by_hour_origin.shape)
total_dep_delays_by_hour_origin.head()
```

Exercise 5: Median Arrival Delay by Carrier and Month
Calculate the median arrival delay for each combination of carrier and month.



```python
# @title
median_arr_del = flights.groupby(['carrier', 'month'])['arr_delay'].median().reset_index()
print(median_arr_del.shape)
median_arr_del.head()
```

Exercise 6: Sort and Display Top Airlines with Most Delayed Flights by Month
For each month, find the airline with the highest average arrival delay, and display the top airline for each month. Filter out months with no delays.



```python
# @title
# Calculate the average arrival delay by airline and month
avg_arr_delay_by_airline_month = flights.groupby(['carrier', 'month'])['arr_delay'].mean().reset_index()

# Sort by average arrival delay in descending order for each month
sorted_by_month = avg_arr_delay_by_airline_month.sort_values(by=['month', 'arr_delay'], ascending=[True, False])

# Filter out months with no delays
filtered_sorted_by_month = sorted_by_month[sorted_by_month['arr_delay'] > 0]

# Display the top airline for each month
top_airlines_by_month = filtered_sorted_by_month.groupby('month').head(1)
```

```python
top_airlines_by_month
```

Exercise 7: Sort and Display Top Origin-Destination Pairs with Highest Average Airtime
Find the top 3 origin-destination pairs with the highest average airtime and display them.

```python
# @title
# Calculate the average airtime for each origin-destination pair
avg_airtime_by_origin_dest = flights.groupby(['origin', 'dest'])['air_time'].mean().reset_index()

# Sort by average airtime in descending order
sorted_by_airtime = avg_airtime_by_origin_dest.sort_values(by='air_time', ascending=False)

# Display the top 3 origin-destination pairs with the highest average airtime
top_origin_dest_pairs = sorted_by_airtime.head(3)
```
